# LINE PLOT

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Buat data dummy

x = np.linspace(0,5,11)
y = x ** 2
```

## Basic Matplotlib: Scripting Layer

```python
plt.figure(figsize=(12,7)) # ini membuat kertas kosong
plt.plot(x,y,'r') # (sumbu x, sumbu y, warna)
plt.xlabel('ini x', size=15)
plt.xticks([1,2,3,4,5],['Andi','Budi','Caca','Dedi','Edi']) # (posisi angka, namanya)
plt.ylabel('ini y', size=15)
plt.yticks([1,2,3,4,5])
plt.title('INI JUDUL', size=20)
plt.show()
```

```python
# buat 2 plot dlm 1 grafik
plt.plot(x,y,'r--')
plt.plot(y,x,'g*')
plt.show()
```

```python
plt.figure(figsize=(12,8))

# subplot untuk buat beberapa grafik dlm 1 figure
plt.subplot(121) # (baris kolom posisi) baris kolomnya maksudnya 1 baris 2 grafik
plt.plot(x,y,'r--')

plt.subplot(122) 
plt.plot(y,x,'g*')

plt.show()
```

```python
plt.figure(figsize=(12,12))

plt.subplot(221) # (baris kolom posisi) baris kolomnya maksudnya 2 baris 2 grafik
plt.plot(x,y,'r--')
plt.xlabel('ini x1')
plt.ylabel('ini y1')
plt.title('ini judul 1')

plt.subplot(222) 
plt.plot(y,x,'g*')
plt.xlabel('ini x1')
plt.ylabel('ini y1')
plt.title('ini judul 2')

plt.subplot(223) 
plt.plot(x,y,'b')
plt.xlabel('ini x1')
plt.ylabel('ini y1')
plt.title('ini judul 3')

plt.suptitle('INI JUDUL UTAMA')

plt.show()
```

## ARTIST LAYER

```python
# buat kanvas/figure kosong
fig = plt.figure(figsize=(10,10))

# tambah axis
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # posisi axis (margin kiri, margin bawah, lebar, tinggi(float 0-1))

# buat plot
axes.plot(x,y,'b--')
axes.set_xlabel('ini sumbu x')
axes.set_ylabel('ini sumbu y')
axes.set_title('INI JUDUL')

plt.show()
```

```python
# buat grafik di dalam grafik

fig = plt.figure(facecolor='yellow')

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b*')
axes1.set_xlabel('ini sumbu x1')
axes1.set_ylabel('ini sumbu y1')
axes1.set_title('INI JUDUL AXES 1')

axes2.plot(y,x,'r--')
axes2.set_xlabel('ini sumbu x2')
axes2.set_ylabel('ini sumbu y2')
axes2.set_title('INI JUDUL AXES 2')
# axes2.set_axis_off() # untuk menghapus axisnya / kotak dan sumbu x dan y
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,8), facecolor='yellow')

for ax in axes:
    # ax.plot(x,y,'b')
    ax.set_xlabel('ini x')
    ax.set_ylabel('ini y')
    ax.set_title('INI JUDUL')

# axes # berisi array posisi subplot
axes[0].plot(x,y)
axes[1].plot(y,x,'r--')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,8), facecolor='pink')

for i in axes:
    for ax in i:
        # ax.plot(x,y,'b')
        ax.set_xlabel('ini x')
        ax.set_ylabel('ini y')
        ax.set_title('INI JUDUL')

# mengakses subplot dgn indexing axes
axes[0][0].plot(x,y,'r--', label='ini 1')
axes[0][0].plot(y,x,'g*', label='ini 2')
axes[0][0].legend()

plt.tight_layout() # merapihkan layout
```

# Saham BANK

```python
bbni = pd.read_csv('BBNI.JK.csv')
bbri = pd.read_csv('BBRI.JK.csv')
bmri = pd.read_csv('BMRI.JK.csv')
btpn = pd.read_csv('BTPN.JK.csv')
bbca = pd.read_csv('BBCA.JK.csv')
```

```python
bbni.head()
```

```python
bbni.info() # cek Null Value dan dtype
```

```python
# ubah date dari str ke datetime
bbni['Date'] = pd.to_datetime(bbni['Date'])
```

```python
bbni.info() # sudah berubah jadi datetime
```

```python
x = bbni['Date']
y1 = bbni['Open']
y2 = bbri['Open']
y3 = bmri['Open']
y4 = btpn['Open']
y5 = bbca['Open']
```

```python
fig = plt.figure(figsize=(20,8))
plt.style.use('ggplot')


bank1 = fig.add_subplot(231)
bank1.plot(x,y1, 'orange')
bank1.set_title('BNI')

bank2 = fig.add_subplot(232)
bank2.plot(x,y2, 'blue')
bank2.set_title('BRI')

bank3 = fig.add_subplot(233)
bank3.plot(x,y3, 'green')
bank3.set_title('Mandiri')

bank4 = fig.add_subplot(234)
bank4.plot(x,y4, 'red')
bank4.set_title('BTPN')

bank5 = fig.add_subplot(235)
bank5.plot(x,y5, 'purple')
bank5.set_title('BCA')

fig.suptitle('PERGERAKAN HARGA SAHAM BANK TAHUN 2020', size=20)

plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.style.use('seaborn')


plt.plot(x,y1,'orange', label='BBNI')
plt.plot(x,y2,'blue', label='BBRI')
plt.plot(x,y3,'green', label='BMRI')
plt.plot(x,y4,'red', label='BTPN')
plt.plot(x,y5,'purple', label='BBCA')

plt.title('PERGERAKAN HARGA SAHAM BANK TAHUN 2020', size=20)
plt.legend(loc=0)
plt.show()
```

# Saham Properti

```python
apln = pd.read_csv('APLN.JK.csv')
asri = pd.read_csv('ASRI.JK.csv')
bsde = pd.read_csv('BSDE.JK.csv')
ctra = pd.read_csv('CTRA.JK.csv')
```

```python
apln.head()
```

```python
apln['Date'] = pd.to_datetime(apln['Date'])
```

```python
apln.info()
```

```python
x = apln['Date']
y1 = apln['Open']
y2 = asri['Open']
y3 = bsde['Open']
y4 = ctra['Open']
```

```python
fig, prop = plt.subplots(nrows=2, ncols=2, figsize=(18,14), facecolor='white')

# plt.style.use('seaborn')
plt.suptitle('PERGERAKAN HARGA SAHAM PROPERTI 2020', size=20)
# fig.autofmt_xdate(rotation=45)

for i in prop:
    for ax in i:
        ax.set_xlabel('Bulan')
        ax.tick_params(axis='x', labelrotation=45) # atur tick sumbu x
        ax.set_ylabel('Harga Saham')

prop[0][0].plot(x,y1,'b')
prop[0][0].set_title('APLN Agung Podomoro')

prop[0][1].plot(x,y2,'r')
prop[0][1].set_title('ASRI Alam Sutra')

prop[1][0].plot(x,y3,'k')
prop[1][0].set_title('BSDE BSD')

prop[1][1].plot(x,y4,'g')
prop[1][1].set_title('CTRA Citra')

plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.style.use('seaborn')


plt.plot(x,y1,'blue', label='APLN')
plt.plot(x,y2,'red', label='ASRI')
plt.plot(x,y3,'black', label='BSDE')
plt.plot(x,y4,'green', label='CTRA')


plt.title('PERGERAKAN HARGA SAHAM PROPERTI TAHUN 2020', size=20)
plt.legend(loc=0)
plt.show()
```

# STACKPLOT

```python
x = cleo['Date']

# menggabungkan data secara vertikal
y = np.vstack([cleo['Open'],myor['Open'],ultj['Open'],icbp['Open']])

y
```

```python
# membuat stackplot untuk percentage data

plt.figure(figsize=(15,8))
plt.style.use('ggplot')

label = ['CLEO', 'MYOR', 'ULTJ', 'ICBP']
color = ['hotpink','royalblue','gold','lime']
plt.stackplot(x,y, labels=label, colors=color, edgecolor='black')

plt.legend(loc=6)
plt.title('SAHAM F&B')
plt.show()
```

```python
x=cleo['Date']

y1 = cleo['Open'].divide(cleo['Open'].sum())
y2 = ultj['Open'].divide(ultj['Open'].sum())
y3 = myor['Open'].divide(myor['Open'].sum())
y4 = icbp['Open'].divide(icbp['Open'].sum())

label = ['CLEO', 'MYOR', 'ULTJ', 'ICBP']
color = ['hotpink','royalblue','gold','lime']

plt.figure(figsize=(10,7))
plt.stackplot(x,y1,y2,y3,y4,labels=label, colors=color, edgecolor='black')

plt.legend(loc=0)
plt.show()
```

# STEM PLOT

```python
x = cleo['Date']
y = cleo['Open'].diff()
# selisih antr data dalam kolom
```

```python
plt.figure(figsize=(10,5))
plt.stem(x,y, markerfmt='b*', linefmt='r--', basefmt='k:', use_line_collection=True)
plt.title('Pergerakan Harga Saham Harian CLEO')
plt.show()
```

```python
x = cleo['Date']
yf1 = cleo['Open'].diff()
yf2 = myor['Open'].diff()
yf3 = ultj['Open'].diff()
yf4 = icbp['Open'].diff()
```

```python
fig, fnb = plt.subplots(ncols=1, nrows=4, figsize=(10,20), facecolor='white')

fnb[0].stem(x,yf1, markerfmt='b*', linefmt='r--', basefmt='k:', use_line_collection=False)
fnb[0].set_title('CLEO')

fnb[1].stem(x,yf2, markerfmt='b*', linefmt='k--', basefmt='k:', use_line_collection=False)
fnb[1].set_title('MYOR')

fnb[2].stem(x,yf3, markerfmt='b*', linefmt='y--', basefmt='k:', use_line_collection=False)
fnb[2].set_title('ULTJ')

fnb[3].stem(x,yf4, markerfmt='b*', linefmt='g--', basefmt='k:', use_line_collection=False)
fnb[3].set_title('ICBP')

plt.suptitle('Pergerakan Harga Saham F&B 2020', size=20)
plt.show()
```

```python
x = apln['Date']
yp1 = apln['Open']
yp2 = asri['Open']
yp3 = bsde['Open']
yp4 = ctra['Open']

plt.figure(figsize=(15,5))

label = ['APLN','ASRI','BSDE','CTRA']
color = ['r','g','b','y']
plt.stackplot(x,yp1,yp2,yp3,yp4, labels=label, colors=color, edgecolor='k')
plt.xticks(rotation=90)
plt.legend(loc=9)
plt.title('STACKPLOT SAHAM PROPERTI 2020')
plt.show()
```
